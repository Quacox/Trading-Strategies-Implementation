```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
S0 = 50  # Current stock price
K = 50  # Strike price
T_short = 2/12  # Time to expiration for the short call (1 month)
T_long = 12/12  # Time to expiration for the long call (3 months)
sigma = 0.2  # Volatility (20%)
r = 0.03  # Risk-free rate (1%)
D = 2  # Net premium paid

# Range of underlying prices at expiration
ST = np.linspace(30, 70, 500)

# Payoff for the short call at expiration
short_call_payoff = np.maximum(ST - K, 0)

# Black-Scholes formula for call option price
def bs_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

# Value of the long call at the expiration of the short call
long_call_value = bs_call(ST, K, T_long - T_short, r, sigma)

# Total PnL
PnL = long_call_value - short_call_payoff - D

# Plot the PnL diagram
plt.figure(figsize=(10, 6))
plt.plot(ST, PnL, label='Calendar Call Spread PnL', color='green')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(K, color='grey', linestyle='--', linewidth=0.5, label='Strike Price K')
plt.axhline(-D, color='red', linestyle='--', linewidth=0.5, label='Net Premium Paid')
plt.axvline(S0, color='blue', linestyle='--', linewidth=0.5, label='Initial Stock Price')

plt.title('Calendar Call Spread Strategy PnL Diagram')
plt.xlabel('Underlying Price at Expiration')
plt.ylabel('Profit / Loss')
plt.xlim(40, 60)
plt.ylim(-10, 10)
plt.legend()
plt.grid(True)
plt.show()

# Display the calculated values
print(f"The first option has a YYM of {T_short}")
print(f"The first option has a YYM of {T_long}")
print(f"Value of Long Call at T' (V): Variable, calculated using Black-Scholes formula")
print(f"Strike Price (K): {K}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

def long_box(K1, K2, D, S_range):
    payoff = []
    for S in S_range:
        payoff_at_expiration = max(0, K1 - S) - max(0, K2 - S) +  max(0, S - K2) - max(0,S - K1) - 2*D
        payoff.append(payoff_at_expiration)
    return np.array(payoff)

# Parameters
K1 = 110
K2 = 95
D = 5

# Calculate payoffs for Long Box
long_box_payoff = long_box(K1, K2, D, S_range)
f_T_bull_call = np.maximum(0, S_range - K2) - np.maximum(0, S_range - K1) - D
f_T_bear_put = np.maximum(0, K1 - S_range) - np.maximum(0, K2 - S_range) - D

# Plot results for Long Box
plt.figure(figsize=(10, 6))
plt.plot(S_range, long_box_payoff, label='Long Box Condor')
plt.plot(S_range, f_T_bull_call, label='Bull Call Spread')
plt.plot(S_range, f_T_bear_put, label='Bear Put Spread')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(K1, color='grey', linestyle='--')
plt.axvline(K2, color='grey', linestyle='--')
plt.title('Long Box Condor Payoff')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('P&L')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def collar(S_0, K1, K2, H, S_range):
    payoff = []
    for S in S_range:
        payoff_at_expiration = S - S_0 + max(0, K1 - S) - max(0, S - K2) - H
        payoff.append(payoff_at_expiration)
    return np.array(payoff)

# Parameters
K1 = 90
K2 = 110
S_0=100
H = 5
S_range = np.linspace(70, 130, 500)

# Calculate payoffs for Collar
collar_payoff = collar(S_0, K1, K2, H, S_range)

# Plot results for Collar
plt.figure(figsize=(10, 6))
plt.plot(S_range, collar_payoff, label='Collar Butterfly')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(K1, color='grey', linestyle='--')
plt.axvline(K2, color='grey', linestyle='--')
plt.axvline(S_0, color='grey', linestyle='--')
plt.title('Collar Payoff')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('P&L')
plt.legend()
plt.grid(True)
plt.show()
```
